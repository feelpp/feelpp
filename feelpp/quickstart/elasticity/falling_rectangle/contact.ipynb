```python
import pandas as pd
df=pd.read_csv('outputs.csv')
df.columns = df.columns.str.replace(' ', '')
print(df.head())
```

```python
# plot 2nd column with plotly
import plotly.graph_objects as go
#
fig = go.Figure(data=[go.Scatter(x=df['time'], y=df['f_C'], mode='lines+markers', name='f_C')])
fig.add_trace(go.Scatter(x=df['time'], y=df['f_t'], mode='markers', name='f_t'))
fig.update_layout(title='f_C vs time', xaxis_title='time', yaxis_title='f_C')
fig.show()
```

```python
# plot mean_displ_ry
fig = go.Figure(data=[go.Scatter(x=df['time'], y=df['mean_displ_ry'], mode='lines+markers', name='mean_displ_ry')])
# add mean_displ_y
fig.add_trace(go.Scatter(x=df['time'], y=df['mean_displ_y'], mode='markers', name='mean_displ_y'))
fig.update_layout(title='mean_displ_ry vs time', xaxis_title='time', yaxis_title='mean_displ_ry')
fig.show()
```
